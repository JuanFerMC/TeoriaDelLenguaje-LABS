# **1. Autómata Finito Determinista (AFD)**

Un Autómata Finito Determinista (AFD) es un modelo matemático utilizado en teoría de la computación para representar sistemas que pueden encontrarse en un número finito de estados y que cambian de estado en función de los símbolos que reciben como entrada (Gribkoff, 2013).

Segun Kourie *et al*., (2012) los AFD se utilizan para modelar y analizar procesos que dependen de la lectura secuencial de símbolos, como el reconocimiento de patrones en cadenas, el análisis de protocolos, el procesamiento de texto y el control de dispositivos electrónicos. Además, los autómatas finitos pueden aplicarse en diferentes áreas de la informática y la ingeniería, incluyendo:

*   Análisis de protocolos de comunicación
*   Procesamiento de texto
*   Inteligencia artificial en videojuegos
*   Reconocimiento de voz

Debido a su capacidad para representar sistemas basados en estados, los AFD permiten describir de forma clara el comportamiento de sistemas que deben mantener una memoria limitada del estado actual mientras procesan una secuencia de datos (Lucas & Reynolds, 2005).

## **2. Definición formal de un Autómata Finito Determinista**

Formalmente, un **Autómata Finito Determinista (AFD)** se define como una **5-tupla**:

$$
M = (Q, \Sigma, \delta, q_0, F)
$$

donde:

- **Q**: conjunto finito de estados del autómata.  
- **Σ**: alfabeto de entrada, es decir, el conjunto de símbolos que el autómata puede leer.  
- **δ**: función de transición que determina el siguiente estado.  
- **q₀**: estado inicial.  
- **F**: conjunto de estados de aceptación.

La función de transición se define como:

$$
\delta : Q \times \Sigma \rightarrow Q
$$

Esto significa que, dado un **estado actual** y un **símbolo de entrada**, el autómata determina **un único estado siguiente**. Por esta razón, el comportamiento del autómata es **determinista**, ya que no existe ambigüedad en las transiciones.
