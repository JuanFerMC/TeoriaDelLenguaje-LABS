##**1. Teoría de Conjuntos.**
Unión de Conjuntos (A∪B).
Esta operación combina todos los elementos de los conjuntos A y B, sin repetir los elementos que se encuentran en ambos.

* **Ejemplo 1:** Si A={1,2,3} y B={3,4,5}, entonces A∪B={1,2,3,4,5}.
* **Ejemplo 2:** Si A={rojo, azul} y B={azul, verde}, entonces A∪B={rojo, azul,verde}.

Se deberá poder entrar cualquier otro grupo de conjuntos.

#**Solucion:**



```python
def union_conjuntos(conjunto_a, conjunto_b):
    resultado = [] # Se guardan los resultados A∪B

    # Conunto que recorre cada elemento del conjunto A y veriifica que no este en el resultado, sino esta lo agrega.
    for elemento in conjunto_a:
        if elemento not in resultado:
            resultado.append(elemento)

    # Conjunto que recorre el conjunto b y hace los mismo del A.
    for elemento in conjunto_b:
        if elemento not in resultado:
            resultado.append(elemento)

    return resultado
#NOTA: El código no distinque masyúsculas por lo que "Verde" es diferente de "verde"

A = input("Conjunto A (Separado por espacios): ").split()
B = input("Conjunto B (Separado por espacios): ").split()

print("Unión de conjuntos:", union_conjuntos(A, B))
```

## **2. Intersección de Conjuntos (A∩B).**
La intersección crea un nuevo conjunto con los elementos que son comunes a los conjuntos A y B.

* **Ejemplo 3:** Si A={a,b,c,d} y B={c,d,e,f}, entonces A∩B={c,d}.
* **Ejemplo 4:** Si A={2,4,6} y B={1,3,5}, entonces A∩B=∅ (conjunto vacío), ya que no tienen elementos en común.

Se deberá poder entrar cualquier otro grupo de conjuntos.

# **Solucion:**

```python
def interseccionConjuntos(A, B):
    resultado = [] # Se guardan los resultados A∪B
    i = 0
    #Condición, si el tamaño de A o de B es 0 entonces retornará un conjunto vacio
    if len(A) == 0 or len(B) == 0:
      return resultado

    while i < len(A):
        if A[i] in B and not A[i] in resultado:
            resultado.append(A[i])
        i += 1
    return resultado

A = input("Conjunto A (Separado por espacios): ").split()
B = input("Conjunto B (Separado por espacios): ").split()

print("Intersección de conjuntos:", interseccionConjuntos(A, B))
```

## **3. Diferencia de Conjuntos (A∖B o A−B).**
La diferencia entre A y B es el conjunto de todos los elementos que están en A pero no en B.

* **Ejemplo 5:** Si A={10,20,30} y B={20,40,50}, entonces A∖B={10,30}.
* **Ejemplo 6:** Si A={manzana,pera,uva} y B={manzana,pera}, entonces A∖B={uva}.

# **Solucion:**

```python
def diferencia_conjuntos(conjunto_a, conjunto_b):
    resultado = [] # Se guardan los resultados A∖B

    # Conjunto que recorre cada elemento del conjunto A
    for elemento in conjunto_a :
      # Condiciones: Que este en A pero no B.
      if elemento not in conjunto_b:
            resultado.append(elemento)

    return resultado

A = input("Conjunto A (Separado por espacios): ").split()
B = input("Conjunto B (Separado por espacios): ").split()

print("Diferencia de conjuntos:", diferencia_conjuntos(A, B))
```

## **4. Diferencia Simétrica (AΔB).**
Esta operación incluye todos los elementos que están en A o en B, pero no en su intersección. Es decir, los elementos que son únicos en cada conjunto.

* **Ejemplo 7:** Si A={a,b,c} y B={c,d,e}, entonces AΔB={a,b,d,e}.
* **Ejemplo 8:** Si A={1,2,3,4} y B={3,4,5,6}, entonces AΔB={1,2,5,6}.

# **Solucion:**

```python
def diferencia_simetrica(conjunto_a, conjunto_b):
    resultado = [] # Se guardan los resultados AΔB

    # Busqueda de los que estan en A mas no en B
    for elemento in conjunto_a :
      if elemento not in conjunto_b:
        if elemento not in resultado:
          resultado.append(elemento)

    # Busqueda de los que estan en B mas no en A
    for elemento in conjunto_b :
      if elemento not in conjunto_a:
        if elemento not in resultado:
            resultado.append(elemento)

    return resultado

A = input("Conjunto A (Separado por espacios): ").split()
B = input("Conjunto B (Separado por espacios): ").split()

print("Diferencia simétrica de conjuntos:", diferencia_simetrica(A, B))
```

## **5. Complemento de un Conjunto (Ac o A′)..**
El complemento de un conjunto A, con respecto a un conjunto universal U, es el conjunto de todos los elementos en U que no están en A.


* **Ejemplo 9**: Si el conjunto universal U={1,2,3,4,5} y A={1,3}, entonces Ac ={2,4,5}..

# **Solucion:**



```python
def complemento(universo,conjuntoa):
    resultado = []

    for elemento in universo:
      if elemento not in conjuntoa:
        if elemento not in resultado:
          resultado.append(elemento)
    return resultado

U = input("Conjunto universal (Separado por espacios): ").split()
A = input("Conjunto A (Separado por espacios): ").split()

resultado_final = complemento(U, A)
print(f"El complemento (Aᶜ) es: {resultado_final}")
```

## **6. Producto Cartesiano (A×B).**
El producto cartesiano de dos conjuntos A y B es el conjunto de todos los pares ordenados (a,b) donde a es un elemento de A y b es un elemento de B.

* **Ejemplo 10:** Si A={1,2} y B={3,4}, entonces A×B={(1,3),(1,4),(2,3),(2,4)}.
# **Solucion:**


```python
def productoCartesiano(conjA, conjB):
    resultado = []
    i = 0
    if len(conjA) != 0 and len(conjB) != 0 and conjA[0] != "" and conjB[0] != "":
        for a in conjA:
            for b in conjB:
                resultado.append((a, b))
        return resultado

    elif len(conjA) == 0 and len(conjB) == 0:
        resultado.append(())
        return resultado

    elif conjA[0] == "":
        for i in conjB:
            resultado.append(("lambda", i))
        return resultado

    elif conjB[0] == "":
        for i in conjA:
            resultado.append((i, "lambda"))
        return resultado

A = input("Ingrese el conjunto A (separado por espacios): ").split()
B = input("Ingrese el conjunto B (separado por espacios): ").split()

print("Producto cartesiano:", productoCartesiano(A, B))
```


## **7. Estrella de Kleene.**
La estrella de Kleene, o clausura de Kleene, se aplica a un conjunto de cadenas o un lenguaje. Es el conjunto de todas las cadenas que se pueden formar concatenando cero o más cadenas del conjunto original. El caso de cero cadenas siempre produce la cadena vacía, denotada por λ.

* **Ejemplo 11:** Si A={a}, entonces:A∗ ={λ,a,aa,aaa,...}
* **Ejemplo 12:** Si A={0,1}, entonces:A∗ es el conjunto de todas las cadenas binarias finitas. Es decir, A∗ ={λ,0,1,00,01,10,11,000,...}
* **Ejemplo 13:** Si A={ab,c}, entonces: A∗ ={λ,ab,c,abab,abc,cab,cc,...}
# **Solucion:**

```python
#Primera version de la Estrella de Kleene (no sirve bien para 2 simbolos)
def estrellaDeKleene(A, n):
    estrella = []

    try:
        n = int(n)
    except ValueError:
        print("¡Debe ingresar un número válido!")
        return ["Lambda"]

    estrella.append("Lambda")

    if n == 0:
        return estrella

    actuales = A[:]  # A¹
    estrella.extend(actuales)

    for i in range(2, n + 1):
        nuevas = []
        for cadena in actuales:
            for simbolo in A:
                nuevas.append(cadena + simbolo)
        estrella.extend(nuevas)
        actuales = nuevas  # Actualizamos para el siguiente nivel

    return estrella


conjunto = input("Ingrese el conjunto de cadenas (separadas por espacios): ").split()
count = input("Ingrese el número de veces que se repetirá la estrella de Kleene: ")

resultado = estrellaDeKleene(conjunto, count)
print("La estrella de Kleene del conjunto es:", resultado)
```
